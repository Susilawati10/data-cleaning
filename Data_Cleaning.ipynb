#5. Deteksi dan Analisis Outlier
#Pertanyaan: Identifikasi outlier pada kolom `no_of_application`. Definisikan outlier menggunakan metode IQR (Interquartile Range) dan berikan analisis singkattentang temuan tersebut.


```python
df.head(3)
```

```python
sns.boxplot(
    data=df,
    x="no_of_application"
)
```

```python
df["no_of_application"].describe()
```

```python
# IQR = Q3 - Q1
q1 = df["no_of_application"].quantile(0.25)
q3 = df["no_of_application"].quantile(0.75)

iqr = q3 - q1

batas_bawah = q1 - (1.5*iqr)
batas_atas = q3 + (1.5*iqr)
```

```python
condition = (df["no_of_application"] > batas_bawah) & (df["no_of_application"] < batas_atas)

df = df[condition]
```

```python
sns.boxplot(
    data=df,
    x="no_of_application"
)
```

```python
df["no_of_application"].describe()
```
